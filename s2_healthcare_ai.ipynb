```python
from google.colab import userdata
import os
os.environ["OPENAI_API_KEY"] = userdata.get('OPENAI_API_Key')
```

```python
from openai import OpenAI
client = OpenAI()

symptoms = input("Enter your symptoms: ")

prompt = f"""
You are an AI healthcare assistant.

Based on the symptoms provided, generate a structured response in the following format:

1. Possible Conditions (2-3 conditions)
2. Severity Level (Low / Medium / High)
3. Recommended Actions
4. When to See a Doctor

Keep the explanation simple and easy to understand.

Symptoms: {symptoms}

IMPORTANT:
- Do not provide exact medical diagnosis
- Add disclaimer: "This is not medical advice"
"""

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": prompt}]
)

result = response.choices[0].message.content

print("\n--- AI Healthcare Assistant Response ---\n")
print(result)
```
