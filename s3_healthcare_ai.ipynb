```python
from google.colab import userdata
import os
os.environ["OPENAI_API_KEY"] = userdata.get('OPENAI_API_Key')
```

```python
from openai import OpenAI
client = OpenAI()

symptoms = input("Enter your symptoms: ")
def get_severity(symptoms):
    symptoms = symptoms.lower()

    if "chest pain" in symptoms or "breathing difficulty" in symptoms:
        return "High"
    elif "fever" in symptoms or "body pain" in symptoms:
        return "Medium"
    else:
        return "Low"

severity = get_severity(symptoms)
prompt = f"""
You are an AI healthcare assistant.

The system has already predicted the severity as: {severity}

Based on the symptoms, generate a structured response:

1. Possible Conditions (2-3)
2. Severity Level (Use given severity: {severity})
3. Recommended Actions
4. When to See a Doctor

Keep it simple and clear.

Symptoms: {symptoms}

IMPORTANT:
- Do not override severity unless absolutely necessary
- Add disclaimer: "This is not medical advice"
"""

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": prompt}]
)
print(f"\nPredicted Severity (System): {severity}")

result = response.choices[0].message.content

print("\n==============================")
print(" AI HEALTHCARE ASSISTANT ")
print("==============================\n")

print(result)

print("\n==============================")
```
